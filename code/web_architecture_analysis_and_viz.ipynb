This script serves as the analysis for a project investigating the differences in food web architecture across ecosystems. The crux of the analysis involves taking previously published food web data and calculating a series of network metrics which illustrate the architecture of the network itself, and thus lend ecological insights about the webs. 

## Set Up Data and Packages

```python
import networkx as nx
from networkx.algorithms import community
import csv
from operator import itemgetter
import pandas as pd
import numpy as np
import matplotlib.pyplot as pltb
import igraph as ig
```

```python
#load data
AEW01_edges = pd.read_csv('C:/Users/coleb/Documents/GitHub/food-web-architecture-across-ecosystems/data/AEW01_edges.csv')
AEW01_nodes = pd.read_csv('C:/Users/coleb/Documents/GitHub/food-web-architecture-across-ecosystems/data/AEW01_nodes.csv')

with open('C:/Users/coleb/Documents/GitHub/food-web-architecture-across-ecosystems/data/AEW01_nodes.csv') as nodecsv: #open file
    nodereader = csv.reader(nodecsv) #read the csv
    #retrieve data using list comprehension
    nodes = [n for n in nodereader][1:]

node_names = [n[0] for n in nodes] # get a list of only the node names
    
with open('C:/Users/coleb/Documents/GitHub/food-web-architecture-across-ecosystems/data/AEW01_edges.csv') as edgecsv: #open file
    edgereader = csv.reader(edgecsv) #read the csv
    #retrieve data using list comprehension
    edges = [n for n in edgereader][1:]
```

## Create Initial Graph

```python
G_AEW01 = nx.Graph()
G_AEW01.add_nodes_from(node_names)
G_AEW01.add_edges_from(edges)
print(nx.info(G_AEW01)) #check basic info
nx.draw(G_AEW01)
```
